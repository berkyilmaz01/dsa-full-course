#### 1.1 Graphs Formal Definitions

# Graph Definition

A graph is an *ordered* pair \( G = (V, E) \), where:

- \( V \) is a finite set of vertices (or nodes).
- \( E \) is a set of edges (or arcs), which are pairs of vertices.

## Important Discussions About Graphs

- If the pairs in **\( E \)** are unordered, the graph is **undirected**. We denote an edge between vertices \( u \) and \( v \) as \( \{u, v\} \).
- If the pairs are **ordered**, the graph is **directed** (a digraph).

### Example

Consider a set of verticles V = {1,2,3,4}

- An *undirected* edge set could be E = {{1,2}, {1,3}, {2,3}, {3,4}}. Note that {1,2} is the same as {2,1}

- A *directed* edge set could be E = {(1,2), (2,1), (1,3), (3,4), (4,3)}. **Here, (1,2) is distinct from (2,1).**
